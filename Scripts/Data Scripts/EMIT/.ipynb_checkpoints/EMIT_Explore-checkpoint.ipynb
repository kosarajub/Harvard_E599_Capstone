```python
def safe_norm_diff(a, b):
    if pd.isna(a) or pd.isna(b) or (a + b) == 0:
        return np.nan
    return (a - b) / (a + b)


def safe_ratio(a, b):
    if pd.isna(a) or pd.isna(b) or b == 0:
        return np.nan
    return a / b
```

```python
def build_bbox(df_group, pad_deg=0.01):
    lon_min = df_group["longitude"].min() - pad_deg
    lon_max = df_group["longitude"].max() + pad_deg
    lat_min = df_group["latitude"].min() - pad_deg
    lat_max = df_group["latitude"].max() + pad_deg
    return (lon_min, lat_min, lon_max, lat_max)

def search_emit_granules(start_date, end_date, bbox, count=200):
    # Official earthaccess pattern supports short_name, temporal, bounding_box
    results = earthaccess.search_data(
        short_name="EMITL2ARFL",
        version="001",
        cloud_hosted=True,
        temporal=(start_date, end_date),
        bounding_box=bbox,
        count=count
    )
    return results


def download_granule_if_needed(granule, out_dir):
    os.makedirs(out_dir, exist_ok=True)

    # earthaccess granule objects usually expose a download() helper or data_links/access links,
    # but the sponsor code already uses a direct HTTPS session. We'll use earthaccess.download.
    downloaded = earthaccess.download([granule], out_dir)
    if not downloaded:
        return None
    return downloaded[0]
```

```python
def extract_emit_spectrum_for_points(nc_path, df_points):
    ds = emit_xarray(nc_path, ortho=True)

    pts = df_points[["row_id", "longitude", "latitude"]].copy()
    pts = pts.set_index("row_id")
    xp = pts.to_xarray()

    extracted = ds.sel(
        latitude=xp.latitude,
        longitude=xp.longitude,
        method="nearest"
    ).to_dataframe().reset_index()

    # Keep only good wavelengths if available, same idea as sponsor script
    if "good_wavelengths" in extracted.columns:
        extracted = extracted[extracted["good_wavelengths"] == 1].copy()

    return extracted
```

```python
def summarize_emit_features(long_df):
    """
    Convert long-format spectrum into one row per row_id with selected wavelengths and indices.
    """
    rows = []

    for row_id, g in long_df.groupby("row_id"):
        g = g.sort_values("wavelengths")

        # helper: nearest reflectance at target wavelength
        def refl_at(target_nm):
            idx = (g["wavelengths"] - target_nm).abs().idxmin()
            return g.loc[idx, "reflectance"]

        r665 = refl_at(665)
        r705 = refl_at(705)
        r740 = refl_at(740)
        r783 = refl_at(783)
        r850 = refl_at(850)
        r1610 = refl_at(1610)
        r2210 = refl_at(2210)

        ndvi = safe_norm_diff(r850, r665)
        ndre1 = safe_norm_diff(r850, r705)
        ndre2 = safe_norm_diff(r850, r740)
        ndre3 = safe_norm_diff(r850, r783)
        ci_rededge = safe_ratio(r850, r705) - 1 if not pd.isna(safe_ratio(r850, r705)) else np.nan
        nbr = safe_norm_diff(r850, r2210)
        msi = safe_ratio(r1610, r850)

        rows.append({
            "row_id": row_id,
            "EMIT_R665": r665,
            "EMIT_R705": r705,
            "EMIT_R740": r740,
            "EMIT_R783": r783,
            "EMIT_R850": r850,
            "EMIT_R1610": r1610,
            "EMIT_R2210": r2210,
            "EMIT_NDVI": ndvi,
            "EMIT_NDRE1": ndre1,
            "EMIT_NDRE2": ndre2,
            "EMIT_NDRE3": ndre3,
            "EMIT_CIrededge": ci_rededge,
            "EMIT_NBR": nbr,
            "EMIT_MSI": msi
        })

    return pd.DataFrame(rows)
```

```python
def process_emit_date_group(df_group, download_dir):
    start_date = str(df_group["start_date"].iloc[0])
    end_date = str(df_group["end_date"].iloc[0])

    bbox = build_bbox(df_group)

    granules = search_emit_granules(start_date, end_date, bbox)

    base = df_group.copy()
    emit_rows = []

    if not granules:
        # no EMIT coverage for this group
        empty = pd.DataFrame({"row_id": base["row_id"]})
        return base.merge(empty, on="row_id", how="left")

    # Very simple strategy:
    # try all matching granules, extract point spectra, keep the first successful hit per row_id
    collected = []

    for granule in granules:
        try:
            nc_path = download_granule_if_needed(granule, download_dir)
            if nc_path is None:
                continue

            long_df = extract_emit_spectrum_for_points(nc_path, base[["row_id", "longitude", "latitude"]])
            if long_df.empty:
                continue

            feat_df = summarize_emit_features(long_df)

            # keep a granule identifier if available
            if hasattr(granule, "meta") and isinstance(granule.meta, dict):
                feat_df["EMIT_granule"] = granule.meta.get("native-id")
            else:
                feat_df["EMIT_granule"] = np.nan

            collected.append(feat_df)

        except Exception as e:
            print(f"EMIT granule failed: {e}")
            continue

    if not collected:
        empty = pd.DataFrame({"row_id": base["row_id"]})
        return base.merge(empty, on="row_id", how="left")

    all_feats = pd.concat(collected, ignore_index=True)

    # If multiple granules hit the same row_id, keep first non-null result.
    # You can later replace this with a better selection rule.
    all_feats = all_feats.sort_values("row_id").groupby("row_id", as_index=False).first()

    return base.merge(all_feats, on="row_id", how="left")
```

```python
def get_agb_data(csv_file):
    agb_df = pd.read_csv(csv_file)

    # Sentinel and Landsat data has information only from 2017.
    threshold_start = pd.Timestamp('2017-01-01')
    threshold_end   = pd.Timestamp('2017-12-31')

    agb_df['start_date'] = pd.to_datetime(agb_df['start_date'])
    agb_df['end_date']   = pd.to_datetime(agb_df['end_date'])

    mask = agb_df['start_date'] < threshold_start
    agb_df.loc[mask, 'start_date'] = threshold_start
    agb_df.loc[mask, 'end_date']   = threshold_end

    agb_df['start_date'] = agb_df['start_date'].dt.strftime('%Y-%m-%d')
    agb_df['end_date']   = agb_df['end_date'].dt.strftime('%Y-%m-%d')

    return agb_df
```

```python
def run_emit(input_csv, output_csv, download_dir):
    df = get_agb_data(input_csv)
    df = df.dropna(subset=["latitude", "longitude", "start_date", "end_date"]).reset_index(drop=True)
    df["row_id"] = df.index.astype(int)

    results = []
    grouped = df.groupby(['start_date', 'end_date'], sort=False)

    for group_num, ((start_date, end_date), df_group) in enumerate(grouped, start=1):
        print(
            f"Processing EMIT group {group_num}/{len(grouped)} | "
            f"start_date={start_date} end_date={end_date} | rows={len(df_group)}"
        )
        merged_group = process_emit_date_group(df_group, download_dir)
        results.append(merged_group)

    df_out = pd.concat(results, ignore_index=True)
    df_out = df_out.sort_values("row_id").drop(columns=["row_id"])
    df_out.to_csv(output_csv, index=False)
    print(f"Saved EMIT output to: {output_csv}")
```

```python
run_emit(INPUT_CSV, OUTPUT_CSV, DOWNLOAD_DIR)
```
