```python
def nearest_refl(wavelengths, reflectance, target_nm):
    idx = np.abs(wavelengths - target_nm).argmin()
    return reflectance[idx]

def spectral_window(wavelengths, values, low_nm, high_nm):
    mask = (wavelengths >= low_nm) & (wavelengths <= high_nm)
    return wavelengths[mask], values[mask]
```

```python
def compute_emit_features_from_spectrum(wavelengths, reflectance):
    wavelengths = np.asarray(wavelengths, dtype=float)
    reflectance = np.asarray(reflectance, dtype=float)

    # Drop invalid values
    valid = np.isfinite(wavelengths) & np.isfinite(reflectance)
    wavelengths = wavelengths[valid]
    reflectance = reflectance[valid]

    if len(wavelengths) < 5:
        return {}

    # Sort just in case
    order = np.argsort(wavelengths)
    wavelengths = wavelengths[order]
    reflectance = reflectance[order]

    # Key reflectance values
    r490  = nearest_refl(wavelengths, reflectance, 490)
    r560  = nearest_refl(wavelengths, reflectance, 560)
    r665  = nearest_refl(wavelengths, reflectance, 665)
    r705  = nearest_refl(wavelengths, reflectance, 705)
    r740  = nearest_refl(wavelengths, reflectance, 740)
    r783  = nearest_refl(wavelengths, reflectance, 783)
    r850  = nearest_refl(wavelengths, reflectance, 850)
    r1610 = nearest_refl(wavelengths, reflectance, 1610)
    r2210 = nearest_refl(wavelengths, reflectance, 2210)

    # Standard interpretable indices
    ndvi  = safe_norm_diff(r850, r665)
    ndre1 = safe_norm_diff(r850, r705)
    ndre2 = safe_norm_diff(r850, r740)
    ndre3 = safe_norm_diff(r850, r783)
    nbr   = safe_norm_diff(r850, r2210)
    msi   = safe_ratio(r1610, r850)

    ratio      = safe_ratio(r850, r705)
    ci_rededge = ratio - 1 if not pd.isna(ratio) else np.nan

    if pd.isna(r850) or pd.isna(r665) or pd.isna(r490):
        evi = np.nan
    else:
        denom = r850 + 6 * r665 - 7.5 * r490 + 1
        evi   = 2.5 * ((r850 - r665) / denom) if denom != 0 else np.nan

    # First derivative of the full spectrum
    dR = np.gradient(reflectance, wavelengths)

    # Red-edge derivative summaries: 680–800 nm
    wl_re, dR_re = spectral_window(wavelengths, dR, 680, 800)
    wl_ref_re, refl_re = spectral_window(wavelengths, reflectance, 680, 800)

    if len(dR_re) > 0:
        max_rededge_slope = np.max(dR_re)
        mean_rededge_slope = np.mean(dR_re)
        std_rededge_slope = np.std(dR_re)
        peak_rededge_wavelength = wl_re[np.argmax(dR_re)]
    else:
        max_rededge_slope = np.nan
        mean_rededge_slope = np.nan
        std_rededge_slope = np.nan
        peak_rededge_wavelength = np.nan

    # Simple red-edge amplitude in reflectance
    if len(refl_re) > 0:
        rededge_reflectance_range = np.max(refl_re) - np.min(refl_re)
    else:
        rededge_reflectance_range = np.nan

    # SWIR derivative summaries: 1500–1800 and 2000–2300 nm
    wl_swir1, dR_swir1 = spectral_window(wavelengths, dR, 1500, 1800)
    wl_swir2, dR_swir2 = spectral_window(wavelengths, dR, 2000, 2300)

    mean_swir1_slope = np.mean(dR_swir1) if len(dR_swir1) > 0 else np.nan
    std_swir1_slope = np.std(dR_swir1) if len(dR_swir1) > 0 else np.nan
    mean_swir2_slope = np.mean(dR_swir2) if len(dR_swir2) > 0 else np.nan
    std_swir2_slope = np.std(dR_swir2) if len(dR_swir2) > 0 else np.nan

    return {
        # Raw narrowband reflectance
        "EMIT_R490": r490,
        "EMIT_R560": r560,
        "EMIT_R665": r665,
        "EMIT_R705": r705,
        "EMIT_R740": r740,
        "EMIT_R783": r783,
        "EMIT_R850": r850,
        "EMIT_R1610": r1610,
        "EMIT_R2210": r2210,

        # Indices
        "EMIT_NDVI": ndvi,
        "EMIT_NDRE1": ndre1,
        "EMIT_NDRE2": ndre2,
        "EMIT_NDRE3": ndre3,
        "EMIT_CIrededge": ci_rededge,
        "EMIT_EVI": evi,
        "EMIT_NBR": nbr,
        "EMIT_MSI": msi,

        # Red-edge derivative summaries
        "EMIT_max_rededge_slope": max_rededge_slope,
        "EMIT_mean_rededge_slope": mean_rededge_slope,
        "EMIT_std_rededge_slope": std_rededge_slope,
        "EMIT_peak_rededge_wavelength": peak_rededge_wavelength,
        "EMIT_rededge_reflectance_range": rededge_reflectance_range,

        # SWIR derivative summaries
        "EMIT_mean_swir1_slope": mean_swir1_slope,
        "EMIT_std_swir1_slope" : std_swir1_slope,
        "EMIT_mean_swir2_slope": mean_swir2_slope,
        "EMIT_std_swir2_slope" : std_swir2_slope,
    }
```

```python
def summarize_emit_features(long_df):
    rows = []

    for row_id, g in long_df.groupby("row_id"):
        g = g.sort_values("wavelengths")

        features = compute_emit_features_from_spectrum(
            wavelengths=g["wavelengths"].to_numpy(),
            reflectance=g["reflectance"].to_numpy()
        )

        if not features:
            continue

        features["row_id"] = row_id
        rows.append(features)

    return pd.DataFrame(rows)
```
